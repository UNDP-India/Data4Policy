```python
import numpy as np
```

```python
# read file info 
fire=gpd.read_file('./fire.shp')
```

```python
# check the columns
fire.columns
```

```python
fire.head()
```

```python
# the data row about acq_date change to the time 
fire_date=pd.DatetimeIndex(fire['acq_date'])
```

```python
# extract the year
fire_date.year
```

```python
#read LULC file
lulc=pd.read_csv('./lulc.csv',index_col=0)
```

```python
lulc
```

```python
# change LULC to the year (fire happened time)
lulc.index=fire_date.year
```

```python
fire.columns
```

```python
lulc
```

```python
# get the fire point, the related year and the landuse type
fire_crop = []
for i,index in enumerate(lulc.index):
    temp  = lulc.iloc[i]
    try:
        value = temp[str(index)]
        fire_crop.append(value)
    except:
        value =  np.NaN
        fire_crop.append(value)
    
```

```python
fire_crop[:10]
```

```python
lulc['fire_year'] = fire_crop
```

```python
lulc
```

```python
lulc
```

 ### we Assume the firepoint happened at the crop area,  this point is the input inner five years，and do the Kmeans 

```python
true_label = []
for i in range(lulc.shape[0]):
    val = lulc['fire_year'].iloc[i]
    if val==5.0:
        true_label.append(1)
    else:
        true_label.append(0)
```

```python
np.unique(true_label)
```

```python
np.count_nonzero(true_label)
```

```python
lulc['true_label'] = true_label
```

```python
lulc
```

### use kmeans to do the unsupervised model

```python
from sklearn.cluster import KMeans
```

```python
lulc.columns
```

```python
# input five year LULC data，do the kmeans analysis
x = lulc[['2017', '2018', '2019', '2020', '2021']].values
```

```python
x
```

```python
# clustering has two center points，fit stand for cluster
kmeans = KMeans(n_clusters=2, random_state=0).fit(x)
```

```python
kmeans.labels_
```

```python
#After the classification，the label 1 is no the agriculture land, vecause of the defination agriculture land is 1，so we use logical_not to change it
predict=np.logical_not(kmeans.labels_).astype(int)
```

```python
predict
```

```python
# using the model to increase the classification rate
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(lulc['true_label'],predict)
```

### out put json

```python
fire_label={}
for fid,label in zip(fire.index,predict):
    fire_label[fid]=str(label)
```
